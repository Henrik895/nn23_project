```python
import numpy as np
import cv2
from matplotlib import pylab as plt
img = cv2.imread("images/train/img1.jpg")
d0=578
d1=770
with open("unit_labels/train/img1.txt") as file_in:
    for line in file_in:
        line_split = line.split()
        x = float(line_split[1])*d1
        y = float(line_split[2])*d0
        w = float(line_split[3])*d1
        h = float(line_split[4])*d0
        x1 = int(x-w/2)
        y1 = int(y-h/2)
        x2 = int(x1+int(w))
        y2 = int(y1+int(h))
        img = cv2.rectangle(img, (x1,y1), (x2,y2), (255,0,0), 2)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
